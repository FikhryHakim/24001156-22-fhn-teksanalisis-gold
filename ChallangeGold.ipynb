# Beri nama kolom

```python
df.columns = ['slang', 'formal']

# Tampilkan beberapa baris pertama setelah penamaan kolom
df.head()
```

```python
slang_dict = dict(zip(df['slang'], df['formal']))
```

```python
def cleanse_text(text, slang_dict):
    sorted_slangs = sorted(slang_dict, key=len, reverse=True)
    pattern = re.compile(r'\b(' + '|'.join(map(re.escape, sorted_slangs)) + r')\b')
    return pattern.sub(lambda x: slang_dict[x.group()], text)
```

```python
# Dataset contoh
data = {
    'text': [
        "pakcikdahtua pakcikmudalagi t3tapjokowi 3x aamiin",
        "nggakpapa saya suka pakcikmudalagi",
        "aamiin saya akan t3tapjokowi"
    ]
}

# Membuat DataFrame dari dataset contoh
df_text = pd.DataFrame(data)

# Terapkan fungsi cleansing pada kolom 'text'
df_text['cleaned_text'] = df_text['text'].apply(lambda x: cleanse_text(x, slang_dict))

# Tampilkan dataset yang telah dibersihkan
df_text
```

```python
# Membaca file data.csv dengan encoding latin1
data_df = pd.read_csv('/content/data.csv', encoding='latin1')
# Membaca file abusive.csv dengan encoding latin1
abusive_df = pd.read_csv('/content/abusive.csv', encoding='latin1')

# Menampilkan beberapa baris pertama dari kedua file untuk mendapatkan gambaran data
data_df.head(), abusive_df.head()
```

```python
# Menghapus baris duplikat
data_df.drop_duplicates(inplace=True)

# Mengatasi data yang hilang dengan menghapus baris yang memiliki nilai NaN
data_df.dropna(inplace=True)

# Menampilkan jumlah baris dan kolom setelah penghapusan
data_df.shape
```

```python
# Mengubah kata-kata abusive menjadi list
abusive_words = abusive_df['ABUSIVE'].tolist()

# Menghapus tweet yang mengandung kata-kata abusive
data_df = data_df[~data_df['Tweet'].str.contains('|'.join(abusive_words), case=False, na=False)]

# Menampilkan jumlah baris dan kolom setelah penghapusan
data_df.shape
```

```python
# Menyimpan dataframe yang telah dibersihkan ke file baru
data_df.to_csv('/content/data.csv', index=False)
```
